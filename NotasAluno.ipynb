```python
# Importando a Biblioteca do Pandas
import pandas as pd
import numpy as np
```

```python
# criando o DataFrame
alunos = pd.DataFrame(data={'Nome':['Aluno 1'],
                            'Sobre Nome':['Sobre Nome'],
                            'Turno':['Manha'],
                             'Serie':['6º Ano'],
                             'Materia':['Matematica'],
                             'Nota':[6],
                             'Status':['Aprovado'],
                             'Mes':['Fevereiro'],
                             'Ano':[2023],
                            },)
```

```python
# Exibir os dados em um DataFrame
alunos
```

```python
# Mudar a Posição da coluna Status 
alunos = alunos[['Nome',	'Sobre Nome',	'Turno',	'Serie',	'Materia',	'Nota',	'Mes','Ano', 'Status']]
```

```python
# Exibindo DataFrame Atualizado
alunos.head()
```

```python
# Criando um novo registro no DataFrame
alunos.loc[len(alunos)] = ['Deik', 'Venus', 'Tarde', '5ºAno','Artes',5, 'Fevereiro', 2023, 'Aprovado']
```

```python
# Exibindo DataFrame com a linha nova
alunos.head(20)
```

```python
# Alterando o nome da coluna Nota para o plural Notas
alunos = alunos.rename(columns={'Nota':'Notas'})
```

```python
# Exibindo DataFrame com a coluna alterada
alunos.head()
```

```python
# Removendo uma Linha do DataFrame
alunos.drop(0, axis=0, inplace=True)
alunos.head()
```
